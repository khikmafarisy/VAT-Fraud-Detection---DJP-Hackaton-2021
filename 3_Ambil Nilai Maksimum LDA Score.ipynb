# Import Libraries

```python
import pandas as pd
import numpy as np
from ast import literal_eval
from operator import itemgetter
```

# Import Hasil Scoring LDA

```python
data_scoring = pd.read_csv('Data Hasil Scoring Nama Barang semua NPWP 1-5 per masa pajak.csv')
```

# VERSION 1

## 1. Terdapat 5 topik dengan 5 skor yang berbeda-beda
## 2. Pada versi 1 ini, nilai skor LDA diambil dari nilai maximum diantara semua topik
## Asumsi : Semua Topik mengenai industri kayu dan turunannya, jika nilai semua topik ~0 maka document tidak berhubungan dengan kayu

```python
data_scoring.head()
```

```python
from tqdm import tqdm
NPWP = []
score_1 = []
score_2 = []
score_3 = []
score_4 = []
score_5 = []
score_6 = []
score_7 = []
score_8 = []
score_9 = []
score_10 = []
score_11 = []
score_12 = []

#for i in tqdm(range(data_scoring.shape[0])):
for i in tqdm(range(data_scoring.shape[0])):
    NPWP.append(data_scoring.NPWP.iloc[i])
    
    if data_scoring.score_1.iloc[i] == '0':
        score_1.append(0)
    else:
        hasil_1 = literal_eval(data_scoring.score_1.iloc[i])
        score_1.append(max(hasil_1,key=itemgetter(1))[1])
    
    if data_scoring.score_2.iloc[i] == '0':
        score_2.append(0)
    else:
        hasil_2 = literal_eval(data_scoring.score_2.iloc[i])
        score_2.append(max(hasil_2,key=itemgetter(1))[1])
    
    if data_scoring.score_3.iloc[i] == '0':
        score_3.append(0)
    else:
        hasil_3 = literal_eval(data_scoring.score_3.iloc[i])
        score_3.append(max(hasil_3,key=itemgetter(1))[1])
    
    if data_scoring.score_4.iloc[i] == '0':
        score_4.append(0)
    else:
        hasil_4 = literal_eval(data_scoring.score_4.iloc[i])
        score_4.append(max(hasil_4,key=itemgetter(1))[1])
    
    if data_scoring.score_5.iloc[i] == '0':
        score_5.append(0)
    else:
        hasil_5 = literal_eval(data_scoring.score_5.iloc[i])
        score_5.append(max(hasil_5,key=itemgetter(1))[1])
    
    if data_scoring.score_6.iloc[i] == '0':
        score_6.append(0)
    else:
        hasil_6 = literal_eval(data_scoring.score_6.iloc[i])
        score_6.append(max(hasil_6,key=itemgetter(1))[1])
    
    if data_scoring.score_7.iloc[i] == '0':
        score_7.append(0)
    else:
        hasil_7 = literal_eval(data_scoring.score_7.iloc[i])
        score_7.append(max(hasil_7,key=itemgetter(1))[1])
    
    if data_scoring.score_8.iloc[i] == '0':
        score_8.append(0)
    else:
        hasil_8 = literal_eval(data_scoring.score_8.iloc[i])
        score_8.append(max(hasil_8,key=itemgetter(1))[1])
    
    if data_scoring.score_9.iloc[i] == '0':
        score_9.append(0)
    else:
        hasil_9 = literal_eval(data_scoring.score_9.iloc[i])
        score_9.append(max(hasil_9,key=itemgetter(1))[1])
    
    if data_scoring.score_10.iloc[i] == '0':
        score_10.append(0)
    else:
        hasil_10 = literal_eval(data_scoring.score_10.iloc[i])
        score_10.append(max(hasil_10,key=itemgetter(1))[1])
    
    if data_scoring.score_11.iloc[i] == '0':
        score_11.append(0)
    else:
        hasil_11 = literal_eval(data_scoring.score_11.iloc[i])
        score_11.append(max(hasil_11,key=itemgetter(1))[1])
    
    if data_scoring.score_12.iloc[i] == '0':
        score_12.append(0)
    else:
        hasil_12 = literal_eval(data_scoring.score_12.iloc[i])
        score_12.append(max(hasil_12,key=itemgetter(1))[1])

pd_cust_max_scoring = pd.DataFrame()
pd_cust_max_scoring['NPWP'] = NPWP

pd_cust_max_scoring['score_1'] = score_1
pd_cust_max_scoring['score_2'] = score_2
pd_cust_max_scoring['score_3'] = score_3
pd_cust_max_scoring['score_4'] = score_4
pd_cust_max_scoring['score_5'] = score_5
pd_cust_max_scoring['score_6'] = score_6
pd_cust_max_scoring['score_7'] = score_7
pd_cust_max_scoring['score_8'] = score_8
pd_cust_max_scoring['score_9'] = score_9
pd_cust_max_scoring['score_10'] = score_10
pd_cust_max_scoring['score_11'] = score_11
pd_cust_max_scoring['score_12'] = score_12
```
